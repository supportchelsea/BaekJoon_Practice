## 10430번
### (A+B)%C는 (A%C + B%C)%C 와 같을까?


```
A,B,C = map(int,input().split())
print((A+B)%C)
print((A%C + B%C)%C)
print((A*B)%C)
print((A%C * B%C)%C)
```

## 2749번
### 피보나치 

재귀함수를 쓸것!!

[풀이과정](https://egg-money.tistory.com/78)



```
# 메모리 에러가 뜬다!!!
num_list = [0,1]
n = int(input())

for i in range(n):
  num_list.append(num_list[i]+num_list[i+1])


sol = num_list[len(num_list)-2]%1000000
print(sol)
```

```
num_list = [0,1]

mod = 10**6
n = int(input())

for i in range(n):
  num_list.append(num_list[i]+num_list[i+1])

num = num_list[n]%mod
print(num)
# 메모리 에러
```

```
mod = 1000000

p = mod//10*15
num = int(input())

listz = [0,1]
for i in range(2,p):
    listz.append(listz[i-1]+listz[i-2])
    listz[i] = listz[i]%mod
print(listz[num%p])

# 이건 왜 이렇게 되는 것인지 잘 모르겠슘...ㅠ
```
