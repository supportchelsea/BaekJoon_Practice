## 9251번 LCS

```
str1 = list(input())
str2 = list(input())
```

```
# 비교 함수
def bigger(comp1, comp2):
    if comp1 > comp2:
        return comp1
    else:
        return comp2
```

```
# 빈껍데기 배열이 필요
# 1칸씩 더 생성한 이유는 나중에 거꾸로 탐색해서 문자를 뽑아내기 위해서야!
# 0이 나타나면 다뽑았다는 것을 알기 위해서!
# 당장은 느낌을 잘 모를 수 있으니 그냥 1칸씩 넉넉하게 더 생성한다고 느끼고 넘어가도 좋아!
# 보통 배열의 인덱스 번호를 1번부터 쓰고싶어할 때도 1칸 늘려서 사용하곤 해
####################################################################

# 검사를 해야해서 check의 약자로 chk라고 변수명을 지었어
# 단어의 순서별로 같은지 아닌지
chk =[[0 for j in range(len(str1)+1)] for i in range(len(str2)+1)]
```

```

# 배열 사이즈 확인
# print(len(chk), len(chk[0]))

# 반복하며 검사하는 알고리즘
# LCS는 특정 인덱스 순서 이후로 얼마나 많은 문자가 겹치는지 확인하는 것!
# 그래서 앞에서 겹친건 중요하지 않고 뒤에서 순서에 맞게 얼마나 겹쳤는지를 확인해야 함
# 위에서 만든 chk는 해당 위치까지 몇개가 겹쳤었는지 누적해서 더해나가는 과정을 나타내기 위해 사용하는 2차원 배열이야
# 2차원 배열인 이유는 2개의 문자열을 비교하기 위함이
####################################################################
# 규칙은 아래와 같아.
# 같은 두 문자에서 str2의 i 번째 문자와 str1의 j 번째 문자가 같다면
# 왼쪽 대각선 위의 값에 1 더한 값을 넣게 돼
# 왜 왼쪽 대각선 위의 값에 1을 더한 값을 넣게 되냐면
# 우선 왼쪽부터 접근하고 있으니까 왼쪽의 값이 더 이전을 의미하는 것이고
# 대각선인 이유는 현재 문자를 비교하기 전까지 마지막으로 누적된 값이 얼마였는지 알기 위해서야

for i in range(1, len(str2)+1):
    for j in range(1, len(str1)+1):
        # 디버그용 확인 출력
        # print(str1[i], str2[j])

        # 위에서 chk를 만들 때 정한 순서가 중요해
        # 현재 i와 j는 chk의 (1,1)부터 값을 채우기 위해서 1부터 시작헀어
        # 그래서 문자열에 값을 비교하기 위해서는 -1을 해줘야 정확한 위치의 문자를 가져올 수 있어
        if str2[i-1] == str1[j-1]:
            chk[i][j] = chk[i-1][j-1]+1
        else:
            chk[i][j] = bigger(chk[i-1][j], chk[i][j-1])
```

```

# 검증용 출력
# print(chk)

# 정답 출력
# 가장 우측하단의 값이 마지막 문자까지 비교했을 때의 누적 값이니까 우측하단 값을 출력하게 되는거야!
print(chk[len(chk)-1][len(chk[0])-1])
```
